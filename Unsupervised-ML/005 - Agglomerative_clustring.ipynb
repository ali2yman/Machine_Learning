```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering 
import seaborn as sns   
```

```python
df = pd.read_csv(r"C:\Users\aliay\OneDrive\Desktop\iTi\21 - ML2\K-means-GMM\Mall_Customers.csv")
df.drop("CustomerID",axis=1,inplace=True)
```

```python
df = pd.get_dummies(df,drop_first=True) 
df.head()
```

```python
# we will find that the relation between Annual Income (k$) and Spending Score (1-100)  the most one to separate the data
sns.pairplot(df)
```

```python
model = AgglomerativeClustering(n_clusters=5,linkage='ward')
predictions = model.fit_predict(df)
predictions
```

```python
plt.scatter(df.iloc[:,1],df.iloc[:,2],c=predictions)
```
