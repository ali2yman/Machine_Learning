```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations =  1000
    n = len(x)
    LR = 0.01

    for i in range(iterations):
        y_pred = (m_curr * x) + b_curr
        cost = (1 / n) * sum ([val**2 for val in (y - y_pred )])
        diff_m = -(2/n)*sum(x * (y - y_pred ) )    # tfadol theta 1
        diff_b = -(2/n)*sum (y - y_pred )          # tfadol theta 0
        m_curr = m_curr - LR * diff_m 
        b_curr = b_curr - LR * diff_b 

        print (f"m = {m_curr} , b = {b_curr} , cost {cost} , iterations {i}")
        
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,9,10,11,13])
```

```python
gradient_descent(x,y)
```
