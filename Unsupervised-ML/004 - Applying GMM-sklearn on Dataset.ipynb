# Gmm on a small Dataset

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
df = pd.read_csv("Clustering_gmm (2).csv")
df.head()
```

```python
plt.scatter(df["Weight"],df["Height"])  #scatter plot
```

```python
gmm = GaussianMixture(n_components=3, random_state=42,n_init=10)
prediction = gmm.fit_predict(df)
prediction
```

```python
# Visualize the clusters
plt.scatter(df["Weight"],df["Height"],c=prediction)
```
