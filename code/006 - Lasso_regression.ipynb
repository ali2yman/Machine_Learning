```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge  
```

```python
dataset = pd.read_csv(r"C:\Users\aliay\OneDrive\Desktop\Supervised Learning Foundation\Machine_Learning\data\data.txt")
dataset
```

```python
df = dataset.copy()
```

```python
X = df.iloc[: ,:-1]
y = df.iloc[: , -1]
```

```python
X
```

```python
y
```

```python
l = .7
ridge = Ridge(alpha = l)
ridge.fit(X,y)
```

```python
ridge_coeff = ridge.coef_
ridge_coeff
```

```python
ridge_intercept = ridge.intercept_
ridge_intercept
```

```python
X_test = [1,2,3,4]
y_pred = ridge.predict(pd.DataFrame(X_test))
y_pred
```

```python
plt.figure(figsize=(12,6) , dpi = 90)
plt.plot(X_test,y_pred , c ='r' , linewidth = 2)
plt.scatter(X,y)
plt.ylim(ymin = 0 , ymax = 9)
plt.xlim(xmin = 0 , xmax = 7)

plt.text(X_test[-1],y_pred[-1] , ' y = '   +    str('%.2f' %ridge_coeff)  + ' * X +'  +  str('%.2f' %ridge_intercept)    
                                                   +  '  For \u03BB or  \u03b1 = ' +    str(l)  ,   fontsize = 12)
plt.title('ridge Regression')
plt.show()
```
